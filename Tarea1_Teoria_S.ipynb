# Definición matemática de momentos iniciales en señales

Sea una señal en el tiempo continuo $x(t)$.  
El k-ésimo momento inicial está definido como:

$$
m_k = \int_{-\infty}^{\infty} t^k \, x(t)\, dt
$$

---

### Señal discreta
Para una señal discreta $(x[n])$, el k-ésimo momento inicial se define como:

$$
m_k = \sum_{n=-\infty}^{\infty} n^k \, x[n]
$$


#  Ejemplo 1: Señal continua

Sea la señal rectangular:

$$
x(t) =
\begin{cases}
1, & 0 \leq t \leq 1 \\
0, & \text{en otro caso}
\end{cases}
$$

---

###  Primer momento inicial

$$
m_1 = \int_{-\infty}^{\infty} t \, x(t)\, dt
$$

Sustituyendo:

$$
m_1 = \int_{0}^{1} t \cdot 1 \, dt = \left[ \frac{t^2}{2} \right]_{0}^{1} = \frac{1}{2}
$$

---

###  Segundo momento inicial

$$
m_2 = \int_{-\infty}^{\infty} t^2 \, x(t)\, dt
$$

Sustituyendo:

$$
m_2 = \int_{0}^{1} t^2 \, dt = \left[ \frac{t^3}{3} \right]_{0}^{1} = \frac{1}{3}
$$

---

 Resultados:

$$
m_1 = \tfrac{1}{2}, \quad m_2 = \tfrac{1}{3}
$$


---

#  Ejemplo 2: Señal discreta

Sea la señal:

$$
x[n] =
\begin{cases}
1, & n = 0,1,2 \\
0, & \text{en otro caso}
\end{cases}
$$

---

###  Primer momento inicial

$$
m_1 = \sum_{n=-\infty}^{\infty} n \cdot x[n]
$$

Sustituyendo:

$$
m_1 = (0 \cdot 1) + (1 \cdot 1) + (2 \cdot 1) = 3
$$

---

###  Segundo momento inicial

$$
m_2 = \sum_{n=-\infty}^{\infty} n^2 \cdot x[n]
$$

Sustituyendo:

$$
m_2 = (0^2 \cdot 1) + (1^2 \cdot 1) + (2^2 \cdot 1) = 0 + 1 + 4 = 5
$$

---

 Resultados:

$$
m_1 = 3, \quad m_2 = 5
$$


#  Definición matemática de momentos centrales

---

###  Señal continua

Sea una señal continua $x(t)$.  
El momento central de orden \(k\) está definido como:

$$
\mu_k = \int_{-\infty}^{\infty} (t - m_1)^k \, x(t)\, dt
$$

donde

$$
m_1 = \int_{-\infty}^{\infty} t \, x(t)\, dt
$$

es el primer momento inicial (valor medio ponderado con $x(t)$.

---

###  Señal discreta

Sea una señal discreta $x[n]$.  
El momento central de orden \(k\) está definido como:

$$
\mu_k = \sum_{n=-\infty}^{\infty} (n - m_1)^k \, x[n]
$$

donde

$$
m_1 = \sum_{n=-\infty}^{\infty} n \cdot x[n]
$$

es el primer momento inicial.



#  Ejemplo 1: Señal continua (rectangular)

Sea la señal:

$$
x(t) =
\begin{cases}
1, & 0 \leq t \leq 1 \\
0, & \text{en otro caso}
\end{cases}
$$

---

###  Valor medio (primer momento inicial)

$$
m_1 = \int_{0}^{1} t \, dt = \left[ \frac{t^2}{2} \right]_{0}^{1} = \frac{1}{2}
$$

---

###  Segundo momento central

Por definición:

$$
\mu_2 = \int_{-\infty}^{\infty} (t - m_1)^2 \, x(t)\, dt
$$

Sustituyendo:

$$
\mu_2 = \int_{0}^{1} (t - \tfrac{1}{2})^2 \, dt
$$

Desarrollando:

$$
= \int_{0}^{1} \left( t^2 - t + \tfrac{1}{4} \right) dt
$$

$$
= \left[ \frac{t^3}{3} - \frac{t^2}{2} + \frac{t}{4} \right]_{0}^{1}
$$

$$
= \frac{1}{3} - \frac{1}{2} + \frac{1}{4} = \frac{1}{12}
$$

---

 Resultado:

$$
\mu_2 = \tfrac{1}{12}
$$


---

#  Ejemplo 2: Señal discreta

Sea la señal:

$$
x[n] =
\begin{cases}
1, & n = 0,1,2 \\
0, & \text{en otro caso}
\end{cases}
$$

---

###  Valor medio (primer momento inicial)

$$
m_1 = \sum_{n=0}^{2} n \cdot x[n] = 0 + 1 + 2 = 3
$$

Como la señal tiene 3 muestras, el valor medio se normaliza (promedio):

$$
m_1 = \frac{3}{3} = 1
$$

---

###  Segundo momento central

Por definición:

$$
\mu_2 = \sum_{n=0}^{2} (n - m_1)^2 \, x[n]
$$

Sustituyendo:

$$
\mu_2 = (0 - 1)^2 \cdot 1 + (1 - 1)^2 \cdot 1 + (2 - 1)^2 \cdot 1
$$

$$
= 1 + 0 + 1 = 2
$$

Normalizando (promedio):

$$
\mu_2 = \frac{2}{3}
$$

---

 Resultado:

$$
\mu_2 = \tfrac{2}{3}
$$


#  Definición matemática del valor medio

---

###  Señal continua

Sea una señal continua $x(t)$.  
El valor medio está definido como:

$$
\bar{x} = \lim_{T \to \infty} \frac{1}{2T} \int_{-T}^{T} x(t) \, dt
$$

---

###  Señal discreta

Sea una señal discreta $x[n]$.  
El valor medio está definido como:

$$
\bar{x} = \lim_{N \to \infty} \frac{1}{2N+1} \sum_{n=-N}^{N} x[n]
$$


#  Ejemplo de valor medio

---

###  Ejemplo 1: Señal continua

Sea la señal periódica cuadrada:

$$
x(t) =
\begin{cases}
1, & 0 \leq t < 1 \\
0, & 1 \leq t < 2
\end{cases}
$$

con periodo \(T = 2\).

---

El valor medio de una señal periódica es:

$$
\bar{x} = \frac{1}{T} \int_{0}^{T} x(t)\, dt
$$

Sustituyendo:

$$
\bar{x} = \frac{1}{2} \int_{0}^{2} x(t)\, dt
$$

$$
= \frac{1}{2} \left( \int_{0}^{1} 1 \, dt + \int_{1}^{2} 0 \, dt \right)
$$

$$
= \frac{1}{2} (1 + 0) = \frac{1}{2}
$$

---

 Resultado:

$$
\bar{x} = \tfrac{1}{2}
$$

---

###  Ejemplo 2: Señal discreta

Sea la secuencia:

$$
x[n] =
\begin{cases}
1, & n = 0,1,2 \\
0, & \text{en otro caso}
\end{cases}
$$

---

El valor medio en \(N=3\) muestras es:

$$
\bar{x} = \frac{1}{3} \sum_{n=0}^{2} x[n]
$$

$$
= \frac{1}{3} (1 + 1 + 1)
$$

$$
= \frac{3}{3} = 1
$$

---

 Resultado:

$$
\bar{x} = 1
$$


#  Definición matemática del valor cuadrático (RMS)

---

###  Señal continua

Sea una señal continua $x(t)$.  
Su valor cuadrático medio (RMS) se define como:

$$
X_{\text{RMS}} = \sqrt{ \lim_{T \to \infty} \frac{1}{2T} \int_{-T}^{T} \big( x(t) \big)^2 \, dt }
$$

---

###  Señal discreta

Sea una señal discreta $x[n]$.  
Su valor cuadrático medio (RMS) se define como:

$$
X_{\text{RMS}} = \sqrt{ \lim_{N \to \infty} \frac{1}{2N+1} \sum_{n=-N}^{N} \big( x[n] \big)^2 }
$$


#  Ejemplo de valor cuadrático (RMS)

---

###  Ejemplo 1: Señal continua (sinusoidal)

Sea la señal:

$$
x(t) = A \sin(\omega t)
$$

con periodo $(T = \frac{2\pi}{\omega})$.

---

El valor cuadrático medio se define como:

$$
X_{\text{RMS}} = \sqrt{\frac{1}{T} \int_{0}^{T} \big(x(t)\big)^2 \, dt}
$$

Sustituyendo:

$$
X_{\text{RMS}} = \sqrt{\frac{1}{T} \int_{0}^{T} A^2 \sin^2(\omega t)\, dt}
$$

Usando la identidad trigonométrica:

$$
\sin^2(\omega t) = \frac{1 - \cos(2\omega t)}{2}
$$

Entonces:

$$
X_{\text{RMS}} = \sqrt{\frac{1}{T} \int_{0}^{T} \frac{A^2}{2} \left( 1 - \cos(2\omega t) \right) dt}
$$

El término de $(\cos(2\omega t))$ se anula al integrar en un periodo, quedando:

$$
X_{\text{RMS}} = \sqrt{\frac{A^2}{2}} = \frac{A}{\sqrt{2}}
$$

---

 Resultado para una señal sinusoidal:

$$
X_{\text{RMS}} = \frac{A}{\sqrt{2}}
$$

---

###  Ejemplo 2: Señal discreta

Sea la secuencia:

$$
x[n] = \{1, \, 2, \, 3\}
$$

---

El valor cuadrático medio es:

$$
X_{\text{RMS}} = \sqrt{\frac{1}{3} \sum_{n=0}^{2} (x[n])^2 }
$$

$$
= \sqrt{\frac{1}{3} \left(1^2 + 2^2 + 3^2\right)}
$$

$$
= \sqrt{\frac{1 + 4 + 9}{3}}
$$

$$
= \sqrt{\frac{14}{3}}
$$

---

 Resultado:

$$
X_{\text{RMS}} = \sqrt{\tfrac{14}{3}} \approx 2.16
$$


#  Definición matemática de la varianza

---

###  Señal continua

Sea una señal continua $x(t)$ con valor medio $(\bar{x})$.  
La varianza se define como:

$$
\sigma^2 = \lim_{T \to \infty} \frac{1}{2T} \int_{-T}^{T} \big( x(t) - \bar{x} \big)^2 \, dt
$$

---

###  Señal discreta

Sea una señal discreta $x[n]$ con valor medio $(\bar{x})$.  
La varianza se define como:

$$
\sigma^2 = \lim_{N \to \infty} \frac{1}{2N+1} \sum_{n=-N}^{N} \big( x[n] - \bar{x} \big)^2
$$


#  Ejemplo de la varianza

---

###  Ejemplo 1: Señal discreta

Sea la secuencia:

$$
x[n] = \{1, 2, 3\}
$$

---

El valor medio es:

$$
\bar{x} = \frac{1}{3} \sum_{n=0}^{2} x[n] = \frac{1+2+3}{3} = 2
$$

La varianza se define como:

$$
\sigma^2 = \frac{1}{3} \sum_{n=0}^{2} \big( x[n] - \bar{x} \big)^2
$$

Sustituyendo valores:

$$
\sigma^2 = \frac{1}{3} \left[ (1-2)^2 + (2-2)^2 + (3-2)^2 \right]
$$

$$
= \frac{1}{3} \left[ 1^2 + 0^2 + 1^2 \right]
$$

$$
= \frac{1+0+1}{3} = \frac{2}{3}
$$

---

 Resultado:

$$
\sigma^2 = \tfrac{2}{3} \approx 0.67
$$

---

###  Ejemplo 2: Señal continua

Sea la señal $x(t)$ en el intervalo [0,1]:

$$
x(t) = t
$$

---

El valor medio es:

$$
\bar{x} = \int_{0}^{1} t \, dt = \left[ \frac{t^2}{2} \right]_{0}^{1} = \frac{1}{2}
$$

La varianza se define como:

$$
\sigma^2 = \int_{0}^{1} \big( t - \tfrac{1}{2} \big)^2 \, dt
$$

Expandiendo:

$$
\sigma^2 = \int_{0}^{1} \left( t^2 - t + \tfrac{1}{4} \right) dt
$$

$$
= \left[ \frac{t^3}{3} - \frac{t^2}{2} + \frac{t}{4} \right]_{0}^{1}
$$

$$
= \frac{1}{3} - \frac{1}{2} + \frac{1}{4}
$$

$$
= \frac{1}{12}
$$

---

 Resultado:

$$
\sigma^2 = \tfrac{1}{12} \approx 0.083
$$


#  Definición matemática de la función de densidad de probabilidad

---

###  Variable aleatoria continua

La función de densidad de probabilidad de una variable aleatoria continua \(X\) se denota por $f_X(x)$ y cumple:

1. **No negatividad**:
$$
f_X(x) \geq 0, \quad \forall x \in \mathbb{R}
$$

2. **Normalización**:
$$
\int_{-\infty}^{\infty} f_X(x) \, dx = 1
$$

3. **Probabilidad en un intervalo**:
$$
P(a \leq X \leq b) = \int_{a}^{b} f_X(x) \, dx
$$

---

###  Variable aleatoria discreta

Para una variable discreta, se define la función de masa de probabilidad $p_X(x)$, que cumple:

1. **No negatividad**:
$$
p_X(x) \geq 0
$$

2. **Normalización**:
$$
\sum_{x \in \mathbb{X}} p_X(x) = 1
$$

3. **Probabilidad de un valor**:
$$
P(X = x) = p_X(x)
$$


#  Ejemplo de función de densidad de probabilidad

---

###  Ejemplo 1: Distribución uniforme continua

Sea una variable aleatoria continua \(X\) con distribución **uniforme** en el intervalo [0,1].  


$$
f_X(x) =
\begin{cases}
1, & 0 \leq x \leq 1 \\
0, & \text{en otro caso}
\end{cases}
$$

---

 Verificación de normalización:

$$
\int_{-\infty}^{\infty} f_X(x)\, dx = \int_{0}^{1} 1 \, dx = 1
$$

 La función está correctamente normalizada.

---

La probabilidad de que \(X\) esté entre \(0.2\) y \(0.6\) es:

$$
P(0.2 \leq X \leq 0.6) = \int_{0.2}^{0.6} 1 \, dx = 0.4
$$

---

###  Ejemplo 2: Distribución normal (Gaussiana)

Sea una variable aleatoria continua \(X\) con distribución normal de media $mu = 0$ y varianza $sigma^2 = 1$.  

Su PDF está dada por:

$$
f_X(x) = \frac{1}{\sqrt{2\pi}} e^{-\frac{x^2}{2}}
$$

---

 Verificación de normalización:

$$
\int_{-\infty}^{\infty} f_X(x) \, dx = 1
$$

---

La probabilidad de que \(X\) esté en el intervalo [-1,1] es:

$$
P(-1 \leq X \leq 1) = \int_{-1}^{1} \frac{1}{\sqrt{2\pi}} e^{-\tfrac{x^2}{2}} \, dx \approx 0.68
$$


#  Definición matemática: Función de densidad marginal (a partir de una conjunta)

---

###  Variables aleatorias continuas

Si \(X\) y \(Y\) son variables aleatorias continuas con **función de densidad conjunta** $f_{X,Y}(x,y)$,  
las **funciones de densidad marginales** se definen como:

- Para \(X\):

$$
f_X(x) = \int_{-\infty}^{\infty} f_{X,Y}(x,y) \, dy
$$

- Para \(Y\):

$$
f_Y(y) = \int_{-\infty}^{\infty} f_{X,Y}(x,y) \, dx
$$

---

###  Variables aleatorias discretas

Si \(X\) y \(Y\) son variables aleatorias discretas con **función de probabilidad conjunta** $p_{X,Y}(x,y)$,  
las **funciones marginales** se definen como:

- Para \(X\):

$$
p_X(x) = \sum_{y} p_{X,Y}(x,y)
$$

- Para \(Y\):

$$
p_Y(y) = \sum_{x} p_{X,Y}(x,y)
$$

---



#  Ejemplo: Función de densidad marginal a partir de una conjunta

---

###  Caso continuo

Sea la función de densidad conjunta de dos variables \(X\) y \(Y\):

$$
f_{X,Y}(x,y) =
\begin{cases}
1, & 0 \leq x \leq 1,\; 0 \leq y \leq 1 \\
0, & \text{en otro caso}
\end{cases}
$$

---

####  Paso 1: Verificar normalización

$$
\int_{-\infty}^{\infty} \int_{-\infty}^{\infty} f_{X,Y}(x,y) \, dx \, dy
= \int_{0}^{1} \int_{0}^{1} 1 \, dx \, dy
= 1
$$

---

####  Paso 2: Calcular la marginal de \(X\)

$$
f_X(x) = \int_{-\infty}^{\infty} f_{X,Y}(x,y) \, dy
= \int_{0}^{1} 1 \, dy =
\begin{cases}
1, & 0 \leq x \leq 1 \\
0, & \text{en otro caso}
\end{cases}
$$

---

####  Paso 3: Calcular la marginal de \(Y\)

$$
f_Y(y) = \int_{-\infty}^{\infty} f_{X,Y}(x,y) \, dx
= \int_{0}^{1} 1 \, dx =
\begin{cases}
1, & 0 \leq y \leq 1 \\
0, & \text{en otro caso}
\end{cases}
$$

---


---

###  Caso discreto

Supongamos que $(X,Y \in \{0,1\})$ con distribución conjunta:

$$
p_{X,Y}(x,y) =
\begin{cases}
\frac{1}{4}, & x,y \in \{0,1\} \\
0, & \text{en otro caso}
\end{cases}
$$

---

- Marginal de \(X\):

$$
p_X(x) = \sum_{y} p_{X,Y}(x,y)
= \frac{1}{4} + \frac{1}{4} = \frac{1}{2}, \quad x \in \{0,1\}
$$

- Marginal de \(Y\):

$$
p_Y(y) = \sum_{x} p_{X,Y}(x,y)
= \frac{1}{4} + \frac{1}{4} = \frac{1}{2}, \quad y \in \{0,1\}
$$

---



#  Definición matemática: Función Generatriz de Momentos (MGF)

---

###  Variable aleatoria continua

Sea \(X\) una variable aleatoria continua con función de densidad $f_X(x)$.  
La **función generatriz de momentos (MGF)** se define como:

$$
M_X(t) = \mathbb{E}[e^{tX}] = \int_{-\infty}^{\infty} e^{t x} f_X(x) \, dx
$$

---

###  Variable aleatoria discreta

Si \(X\) es una variable discreta con función de probabilidad $p_X(x)$, la MGF es:

$$
M_X(t) = \mathbb{E}[e^{tX}] = \sum_{x} e^{t x} \, p_X(x)
$$

---

###  Relación con los momentos

Los momentos de orden \(n\) de \(X\) se obtienen derivando la MGF:

$$
m_n = \mathbb{E}[X^n] = \left. \frac{d^n}{dt^n} M_X(t) \right|_{t=0}
$$


#  Ejemplo: Función Generadora de Momentos (MGF)

---

###  Variable Bernoulli

Sea $(X \sim \text{Bernoulli}(p))$, con función de probabilidad:

$$
p_X(x) =
\begin{cases}
p, & x = 1 \\
1-p, & x = 0
\end{cases}
$$

---

####  Paso 1: Definición de la MGF

$$
M_X(t) = \mathbb{E}[e^{tX}] = \sum_{x} e^{t x} \, p_X(x)
$$

---

####  Paso 2: Sustitución

$$
M_X(t) = e^{t \cdot 0}(1-p) + e^{t \cdot 1}(p)
= (1-p) + p e^t
$$

---

####  Paso 3: Cálculo de momentos a partir de la MGF

- Primer momento (valor esperado):

$$
\mathbb{E}[X] = M_X'(0) = \frac{d}{dt}\Big[ (1-p) + p e^t \Big]_{t=0} = p
$$

- Segundo momento:

$$
\mathbb{E}[X^2] = M_X''(0) = \frac{d^2}{dt^2}\Big[ (1-p) + p e^t \Big]_{t=0} = p
$$

---

 **Conclusión:**  
La MGF de una variable Bernoulli es:

$$
M_X(t) = (1-p) + p e^t
$$

y nos permite obtener fácilmente todos los momentos de \(X\).


#  Definición matemática de un conjunto

Un conjunto es una colección bien definida de elementos, que pueden ser números, símbolos u objetos.

---

###  Notación general

Un conjunto \( A \) se denota como:

$$
A = \{ a_1, a_2, a_3, \dots \}
$$

donde cada elemento $( a_i )$ pertenece al conjunto.

---

###  Ejemplo

El conjunto de números naturales hasta 5:

$$
A = \{ 1,2,3,4,5 \}
$$

El conjunto de números pares puede definirse por una regla de formación:

$$
B = \{ x \in \mathbb{N} \; | \; x \; \text{es par} \}
$$


#  Definición matemática de una métrica

Sea \(X\) un conjunto no vacío.  
Una **métrica** en \(X\) es una función

$$
d : X \times X \to \mathbb{R}
$$

tal que para todo $(x,y,z \in X)$, se cumplen las siguientes propiedades:

1. **No negatividad:**

$$
d(x,y) \geq 0
$$

2. **Identidad del discernible:**

$$
d(x,y) = 0 \iff x = y
$$

3. **Simetría:**

$$
d(x,y) = d(y,x)
$$

4. **Desigualdad triangular:**

$$
d(x,z) \leq d(x,y) + d(y,z)
$$

---

###  Ejemplo: métrica euclidiana en $(\mathbb{R}^n)$

Para $(x=(x_1, \dots, x_n), \; y=(y_1, \dots, y_n) \in \mathbb{R}^n)$:

$$
d(x,y) = \sqrt{\sum_{i=1}^n (x_i - y_i)^2}
$$


#  Ejemplo de métrica: Distancia Manhattan en $(\mathbb{R}^2)$

Sea el conjunto $(X = \mathbb{R}^2)$.  
Definimos la **métrica Manhattan** como:

$$
d(x,y) = |x_1 - y_1| + |x_2 - y_2|,
\quad x=(x_1,x_2), \; y=(y_1,y_2) \in \mathbb{R}^2
$$

---

### ✅ Verificación de propiedades

1. **No negatividad:**

$$
d(x,y) = |x_1-y_1| + |x_2-y_2| \geq 0
$$

2. **Identidad del discernible:**

$$
d(x,y) = 0 \iff |x_1-y_1| = 0 \; \text{y} \; |x_2-y_2| = 0 \iff x=y
$$

3. **Simetría:**

$$
d(x,y) = |x_1-y_1| + |x_2-y_2| = |y_1-x_1| + |y_2-x_2| = d(y,x)
$$

4. **Desigualdad triangular:**

Para $(x,y,z \in \mathbb{R}^2)$:

$$
d(x,z) \leq d(x,y) + d(y,z)
$$

(lo cual se cumple por la desigualdad triangular del valor absoluto).

---

###  Ejemplo numérico

Sean $(x=(1,2))$ y $(y=(4,6))$:

$$
d(x,y) = |1-4| + |2-6| = 3 + 4 = 7
$$


#  Definición matemática de un espacio

Un espacio es una estructura matemática formada por un conjunto junto con propiedades o relaciones adicionales que permiten trabajar con él.  

---

###  Espacio Métrico

Sea \(X\) un conjunto no vacío y $(d : X \times X \to \mathbb{R})$ una métrica en \(X\).  
El par \((X,d)\) se denomina **espacio métrico**, donde:

$$
d(x,y) \geq 0, \quad d(x,y) = 0 \iff x=y, \quad d(x,y) = d(y,x), \quad d(x,z) \leq d(x,y)+d(y,z)
$$

---

###  Espacio Vectorial

Sea $(\mathbb{K})$ un cuerpo (por ejemplo, $(\mathbb{R})$ o $(\mathbb{C})$).  
Un conjunto \(V\) con dos operaciones (suma y producto por un escalar) es un **espacio vectorial** sobre $(\mathbb{K})$ si:

1. $((V,+))$ es un grupo abeliano.  
2. Para todo $(v \in V, \; \alpha \in \mathbb{K})$, se cumple $(\alpha v \in V)$.  
3. Se cumplen los axiomas de distributividad, asociatividad y existencia de neutro.

---

###  Ejemplo

- $((\mathbb{R}^n, d))$ con la métrica euclidiana:

$$
d(x,y) = \sqrt{\sum_{i=1}^n (x_i - y_i)^2}, \quad x,y \in \mathbb{R}^n
$$

es un **espacio métrico**.  

- $(\mathbb{R}^n)$ con las operaciones usuales de suma y producto por escalar es un **espacio vectorial** sobre $(\mathbb{R})$.


#  Definición matemática de un espacio probabilístico

Un **espacio probabilístico** es una terna:

$$
(\Omega, \mathcal{F}, P)
$$

donde:

1. $(\Omega)$ es el **espacio muestral**, es decir, el conjunto de todos los posibles resultados de un experimento aleatorio.

2. $(\mathcal{F}$) es una **σ-álgebra** de subconjuntos de $(\Omega)$, es decir:
   - $(\Omega \in \mathcal{F})$,
   - si $(A \in \mathcal{F})$, entonces $(A^c \in \mathcal{F})$,
   - si $(A_1, A_2, \dots \in \mathcal{F})$, entonces $(\bigcup_{i=1}^\infty A_i \in \mathcal{F})$.

3. $(P : \mathcal{F} \to [0,1])$ es una **medida de probabilidad** que cumple:
   - $(P(\Omega) = 1)$,
   - (σ-aditividad) si $(A_1, A_2, \dots \in \mathcal{F})$ son disjuntos dos a dos, entonces:

     $$
     P\!\left(\bigcup_{i=1}^\infty A_i\right) = \sum_{i=1}^\infty P(A_i)
     $$
---

###  Ejemplo

Lanzamiento de un dado:

- $(\Omega = \{1,2,3,4,5,6\})$  
- $(\mathcal{F} = \mathcal{P}(\Omega))$ (el conjunto de todos los subconjuntos de $(\Omega)$  
- $(P(\{k\}) = \tfrac{1}{6}, \quad k=1,\dots,6)$.


#  Ejemplo de espacio probabilístico

Consideremos el experimento aleatorio de **lanzar una moneda balanceada**.  

El espacio probabilístico está dado por:

$$
(\Omega, \mathcal{F}, P)
$$

donde:

- **Espacio muestral:**
$$
\Omega = \{C, X\}
$$
donde \(C\) representa "cara" y \(X\) representa "cruz".

- **σ-álgebra:**
$$
\mathcal{F} = \{\varnothing, \{C\}, \{X\}, \Omega\}
$$

- **Medida de probabilidad:**
$$
P(\{C\}) = \tfrac{1}{2}, \quad P(\{X\}) = \tfrac{1}{2}, \quad P(\Omega)=1
$$

---

###  Verificación de axiomas de Kolmogórov

1. **No negatividad:**
$$
P(A) \geq 0, \quad \forall A \in \mathcal{F}
$$

2. **Normalización:**
$$
P(\Omega) = 1
$$

3. **Aditividad numerable:**

Si $(A_1 = \{C\})$, $(A_2 = \{X\})$, con $(A_1 \cap A_2 = \varnothing)$:

$$
P(A_1 \cup A_2) = P(\{C,X\}) = 1 = \tfrac{1}{2} + \tfrac{1}{2} = P(A_1)+P(A_2)
$$


#  Definición matemática de una σ-álgebra

Sea $(\Omega)$ un conjunto no vacío.  
Una **σ-álgebra** $(\mathcal{F})$ sobre $(\Omega)$ es una colección de subconjuntos de \(\Omega\) que cumple:

1. $(\Omega \in \mathcal{F})$.
   
2. Si $(A \in \mathcal{F})$, entonces su complemento $(A^c = \Omega \setminus A \in \mathcal{F})$.
   
3. Si $(A_1, A_2, \dots \in \mathcal{F})$, entonces:
   $$
   \bigcup_{i=1}^{\infty} A_i \in \mathcal{F}
   $$

---

En otras palabras:

$$
\mathcal{F} \subseteq \mathcal{P}(\Omega), \quad
\mathcal{F} \; \text{es cerrada bajo complementos y uniones numerables.}
$$


#  Ejemplo de σ-álgebra

Sea el espacio muestral:

$$
\Omega = \{1,2,3\}
$$

Una **σ-álgebra** $(\mathcal{F})$ sobre $(\Omega)$ puede ser:

$$
\mathcal{F} = \big\{ \varnothing, \{1,2,3\}, \{1\}, \{2,3\} \big\}
$$

---

###  Verificación de propiedades

1. **Contiene a $(\Omega)$:**
$$
\Omega = \{1,2,3\} \in \mathcal{F}
$$

2. **Cerrada bajo complemento:**
- El complemento de $(\{1\})$ es $(\{2,3\} \in \mathcal{F})$.  
- El complemento de $(\{2,3\}) es (\{1\} \in \mathcal{F})$.  
- El complemento de $(\Omega) es (\varnothing \in \mathcal{F})$.

3. **Cerrada bajo uniones numerables:**
$$
\{1\} \cup \{2,3\} = \Omega \in \mathcal{F}
$$

---

###  Otros ejemplos de σ-álgebra

- **Trivial:**
$$
\mathcal{F} = \{\varnothing, \Omega\}
$$

- **Completa (potencia):**
$$
\mathcal{F} = \mathcal{P}(\Omega) = \{\varnothing, \{1\}, \{2\}, \{3\}, \{1,2\}, \{1,3\}, \{2,3\}, \Omega\}
$$


#  Definición matemática: Axiomas de Kolmogórov

Sea $((\Omega, \mathcal{F}, P))$ un **espacio probabilístico**, donde:

- $(\Omega)$ es el espacio muestral,
- $(\mathcal{F})$ es una σ-álgebra de subconjuntos de $(\Omega)$,
- $(P : \mathcal{F} \to [0,1])$ es una función de probabilidad.

La probabilidad \(P\) cumple los siguientes **axiomas de Kolmogórov**:

---

### 1. **No negatividad**
$$
P(A) \geq 0, \quad \forall A \in \mathcal{F}
$$

---

### 2. **Normalización**
$$
P(\Omega) = 1
$$

---

### 3. **Aditividad numerable (σ-aditividad)**
Si $(A_1, A_2, A_3, \dots \in \mathcal{F})$ son **disjuntos dos a dos**, es decir

$(A_i \cap A_j = \varnothing)$ para $(i \neq j)$, entonces:
$$
P\!\left(\bigcup_{i=1}^{\infty} A_i\right) = \sum_{i=1}^{\infty} P(A_i)
$$


#  Ejemplo: Axiomas de Kolmogórov

Consideremos el lanzamiento de un dado justo.  
El espacio muestral es:

$$
\Omega = \{1,2,3,4,5,6\}
$$

La σ-álgebra puede ser la **potencia completa**:

$$
\mathcal{F} = \mathcal{P}(\Omega)
$$

La probabilidad \(P\) se define de manera uniforme:

$$
P(\{i\}) = \frac{1}{6}, \quad i = 1,2,3,4,5,6
$$

---

###  Verificación de los axiomas

1. **No negatividad**
$$
P(A) \geq 0, \quad \forall A \in \mathcal{F}
$$
Ejemplo: $(P(\{3\}) = \tfrac{1}{6} \geq 0)$.

---

2. **Normalización**
$$
P(\Omega) = P(\{1,2,3,4,5,6\}) = \sum_{i=1}^{6} P(\{i\}) = 6 \cdot \tfrac{1}{6} = 1
$$

---

3. **Aditividad numerable**
Si tomamos $(A_1 = \{1\}, A_2 = \{2\}, A_3 = \{3\})$, que son **disjuntos**, entonces:

$$
P(A_1 \cup A_2 \cup A_3) = P(\{1,2,3\}) = \frac{3}{6} = \frac{1}{2}
$$

y

$$
P(A_1) + P(A_2) + P(A_3) = \tfrac{1}{6} + \tfrac{1}{6} + \tfrac{1}{6} = \tfrac{1}{2}
$$

 Se cumple la aditividad.
